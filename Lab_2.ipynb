```python
df=pd.read_csv('/content/salary_prediction_data.csv')
df
```

Построим график  распределения зарплат и сравним с графиком нормального распределения

```python
salaries = df['Salary']

plt.hist(salaries, bins=40, density=True, alpha=0.6, color="g", label="Зарплаты в USD")

mean, std = salaries.mean(), salaries.std()

xmin, xmax = plt.xlim()
x = np.linspace(xmin, xmax, 100)
p = norm.pdf(x, mean, std)

plt.plot(x, p, "k", linewidth=2, label="Нормальное распределение")

plt.title("Распределение зарплат в USD")
plt.xlabel("Зарплата (USD)")
plt.ylabel("Плотность")
plt.legend()
plt.show()
```

Рассчитаем математическое ожидание, дисперсию, асимметрию, эксцесс, квантили 0,05 и 0,95 и 2,5%-ную точку

```python
mean_value_2 = df['Salary'].mean()
print(f"Математическое ожидание: {mean_value_2:.3f} USD")

variance_2 = df['Salary'].var()
print(f"Дисперсия: {variance_2:.3f} USD²")

skewness_2 = df['Salary'].skew()
print(f"Асимметрия: {skewness_2:.3f}")

kurtosis_2 = df['Salary'].kurtosis()
print(f"Эксцесс: {kurtosis_2:.3f}")

quantile_05_2 = df['Salary'].quantile(0.05)
quantile_95_2 = df['Salary'].quantile(0.95)
print(f"Квантиль 0,05: {quantile_05_2:.3f} USD")
print(f"Квантиль 0,95: {quantile_95_2:.3f} USD")

quantile_025_2 = df['Salary'].quantile(0.025)
print(f"2,5%-ная точка: {quantile_025_2:.3f} USD")
```
