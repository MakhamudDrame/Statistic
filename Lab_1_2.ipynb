```python
# импортируем библиотеки
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# читаем  файл бмв
bmw=pd.read_csv('/content/uk_used_cars/bmw.csv')
```

```python
bmw.describe()
```

```python
# рассчитаем среднее, медиану и стандартное отклонение сгруппировав автомобили бмв по типу трансмиссии
bmw.groupby('transmission')['price'].agg(['count','mean','median','std'])
```

```python
# Построим график ящик с усами для показателя датафрейма цена  в зависимости от типа трансмиссии
bmw.boxplot(column=['price'], by='transmission', figsize=(15,5), grid = False)
```

```python
bmw.loc[bmw['transmission'] == 'Manual'].hist(column='price', bins=50)
```

```python
bmw.loc[bmw['transmission'] == 'Automatic'].hist(column='price',bins=50)
```

```python
bmw.loc[bmw['transmission'] == 'Semi-Auto'].hist(column='price',bins=50)
```

У поддержанных автомобилей все выбросы идут вверх, возможно это связано с тем, что более новые модели продаются по более высоким цена

```python
# Построим график ящик с усами для показателя датафрейма цена  в зависимости от года выпуска
bmw.boxplot(column=['price'], by='year', figsize=(15,5), grid = False)
```

Действительно, чем новее год выпуска - тем больше выбросов за пределами 0,75 квартиля

```python
df=pd.read_csv('/content/Salary_Data.csv')
```

```python
df
```

```python
#  посчитаем средний доход, медиану и стандартное отклонение в зависимости от уровня образования
df.groupby('Education Level')['Salary'].agg(['count','mean','median','std'])
```

```python
# Приведем к одному названию значения, которые одинаковы по смыслу и написаны по-разному
df['Education Level'].replace({'Bachelor\'s': 'Bachelor\'s Degree', 'Master\'s': 'Master\'s Degree','phD':'PhD'}, inplace=True)
```

```python
#  посчитаем средний доход, медиану и стандартное отклонение в зависимости от уровня образования
df.groupby('Education Level')['Salary'].agg(['count','mean','median','std'])
```

```python
df.describe()
```

```python
# Посчитаем разницу в уровне дохода в зависимости от образования
df.boxplot(column=['Salary'], by='Education Level', figsize=(15,5), grid = False)
```

Мы видим, что чем выше уровень образования, тем выше медианный и средний доход. Особенно доход растет после получения степени бакалавра. При этом выбросы при достижении степени магистра отсутствуют, а при степени бакалавра являются единичными. Другая ситуация при наличии школьного образования и кандидатской степени. При наличии школьного образования все выбросы сосредоточены выше квартиля 0,75, это значит , что не редки случаи, когда люди без высшего образования имеют не меньший доход, чем люди с высшим образованием. При этом кандидатская степень  дает общий прирост дохода, а не гарантирует более высокий доход, так как все выбросы сосредоточены ниже нижнего 0,25 квартиля

```python
# Посчитаем разницу в уровне дохода в зависимости от опыта
df.boxplot(column=['Salary'], by='Years of Experience', figsize=(20,5), grid = False)
```

Мы видим, что при начальном опыте выбросы по доходу в первые 7 лет опыта идут почти всегда только вверх, что связано с тем, что квалификация больше влияет на жохож, чем годы опыта. При достижении  опыта от 12 до 20 лет выбросы по доходу, чаще идут ниже 0,25 квартиля - роль опыта играет все меньшую роль. После 20 лет опыта выборка сильно сокращяется и не дает нам сделать статистически значимый вывод
