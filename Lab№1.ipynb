```python
zp=pd.read_excel('/content/sr-zpl_2023.xls',header=3)
zp
```

```python
zp.rename(columns={'Unnamed: 0':'year','Unnamed: 1':'Зарплата всего','первая \n(с наименьшей заработной платой)':'первая с наименьшей заработной платой',
                   'десятая\n(с наибольшей заработной платой)'	: 'с наибольшей заработной платой'
                  },inplace=True)
zp
```

```python
zp=zp[0:16]
zp
```

```python
x=zp.iloc[15]              # Возьмем последний год наблюдений и посчитаем медиану средних значемий каждого дециля работников
x                          # по доходу  и запишем в отдельную переменную
```

```python
a=x[2:12].median()          # рассчитаем медианную зп для всех децилей
a
```

```python
b=x[2:12].mean()            # рассчитаем среднюю зп для всех децилей
b
```

```python
b/a                         # во сколько раз средняя зп выше медианы
```

В данных по распределению зарплаты между 1 и 10 децилями, равными по количеству людей, мы видим то во сколько раз среднее значение оходов выше медианы. Это связанно с неравномерностью распределения доходов
